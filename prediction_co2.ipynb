### Analyse des caract√©ristiques les plus importantes:

**Gaz (importance: 0.73)**: C'est de loin la caract√©ristique la plus importante, ce qui indique que la production de gaz a une influence majeure sur le taux de CO2. Cela est logique, car la combustion du gaz naturel lib√®re du CO2.

**Charbon (importance: 0.08)**: Le charbon est la deuxi√®me caract√©ristique la plus importante. Comme le gaz, la combustion du charbon est une source majeure d'√©missions de CO2, et son importance est bien refl√©t√©e ici.

**Nucl√©aire (importance: 0.04)**: L'√©nergie nucl√©aire, bien que ne produisant pas directement de CO2 pendant son fonctionnement, peut avoir une influence indirecte sur le mix √©nerg√©tique global et est donc captur√©e comme une caract√©ristique importante par le mod√®le.

**Consommation (importance: 0.02)**: La consommation globale d'√©nergie influence naturellement les √©missions de CO2, car une plus grande demande doit √™tre satisfaite par diverses sources d'√©nergie, y compris celles qui √©mettent du CO2.

**√âolien (importance: 0.018)**: L'√©nergie √©olienne est une source d'√©nergie renouvelable. Son importance n√©gative ou positive (selon le coefficient exact si c'√©tait une r√©gression lin√©aire, ici c'est une importance) peut refl√©ter son r√¥le dans la r√©duction des √©missions de CO2 en rempla√ßant les sources fossiles, ou simplement sa variabilit√© affectant le mix √©nerg√©tique.

**Charbon_lag1 (importance: 0.014)**: La valeur pr√©c√©dente de la production de charbon est √©galement une caract√©ristique importante, ce qui sugg√®re une certaine inertie ou d√©pendance temporelle dans l'utilisation du charbon et, par cons√©quent, dans les √©missions de CO2.

**Ech. physiques (importance: 0.013)**: Cette caract√©ristique fait probablement r√©f√©rence aux √©changes physiques d'√©nergie avec d'autres pays, ce qui peut influencer l'offre et la demande domestiques, et donc les sources d'√©nergie utilis√©es et leurs √©missions de CO2.

**Solaire (importance: 0.009)**: L'√©nergie solaire, une autre source renouvelable, contribue √©galement, bien que de mani√®re moins importante que l'√©olien, √† influencer les √©missions de CO2 par son r√¥le dans le mix √©nerg√©tique.

**Charbon_lag2 (importance: 0.006)**: Similaire √† 'Charbon_lag1', les valeurs ant√©rieures de la production de charbon continuent d'avoir une influence, bien que d√©croissante.

**Gaz - Cog√©n. (importance: 0.005)**: La cog√©n√©ration au gaz est une m√©thode de production d'√©nergie qui peut √™tre plus efficace, et son volume influence √©galement les √©missions de CO2.

```python
from sklearn.ensemble import RandomForestRegressor

df_model = df_clean.copy() # Make a copy to avoid modifying df_clean directly

# cr√©ation des features lag
lags = [1, 2, 3]  # d√©calages
for lag in lags:
    df_model[f'Gaz_lag{lag}'] = df_model['Gaz'].shift(lag)
    df_model[f'Charbon_lag{lag}'] = df_model['Charbon'].shift(lag)

# drop NaN (√† cause du lag)
df_model = df_model.dropna()

X = df_model.drop('Taux de Co2', axis=1)
y = df_model['Taux de Co2']

# Select only numerical features for X, as Random Forest can't handle datetime/object types directly
X_numeric = X.select_dtypes(include=['number'])

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_numeric, y)

importance = pd.Series(rf.feature_importances_, index=X_numeric.columns).sort_values(ascending=False)
print(importance)
```

```python
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import r2_score, mean_squared_error
```

```python
X = X_selected
y = df_model["Taux de Co2"] # y is already defined, but re-assigning for clarity
```

```python
split_index = int(len(X) * 0.8)

X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]
```

```python
tscv = TimeSeriesSplit(n_splits=5)
```

```python
pipeline_ridge = Pipeline([
    ("scaler", StandardScaler()),
    ("model", Ridge())
])

param_grid_ridge = {
    "model__alpha": [0.1, 1, 10, 100]
}

grid_ridge = GridSearchCV(
    pipeline_ridge,
    param_grid_ridge,
    cv=tscv,
    scoring="r2"
)

grid_ridge.fit(X_train, y_train)
```

```python
param_grid_rf = {
    "n_estimators": [100, 200],
    "max_depth": [10, 20, None]
}

grid_rf = GridSearchCV(
    RandomForestRegressor(random_state=42),
    param_grid_rf,
    cv=tscv,
    scoring="r2",
    n_jobs=-1
)

grid_rf.fit(X_train, y_train)
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor(
    objective="reg:squarederror",
    random_state=42
)
param_grid_xgb = {
    "n_estimators": [100, 200, 500],
    "max_depth": [3, 4, 6],
    "learning_rate": [0.01, 0.05, 0.1],
    "subsample": [0.8, 1.0],
    "colsample_bytree": [0.8, 1.0]
}

# üîπ GridSearchCV avec TimeSeriesSplit (tscv d√©j√† d√©fini)
grid_xgb = GridSearchCV(
    xgb,
    param_grid_xgb,
    cv=tscv,
    scoring="r2",
    n_jobs=-1
)

grid_xgb.fit(X_train, y_train)
```

```python
import numpy as np

models = {
    "Ridge": grid_ridge,
    "Random Forest": grid_rf,
    "XGBoost": grid_xgb
}

for name, model in models.items():
    # Pr√©dictions sur TRAIN
    y_train_pred = model.predict(X_train)
    r2_train = r2_score(y_train, y_train_pred)
    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))

    # Pr√©dictions sur TEST
    y_test_pred = model.predict(X_test)
    r2_test = r2_score(y_test, y_test_pred)
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))

    print(f"\n{name}")
    print("Best params :", model.best_params_)
    print("R2 TRAIN :", round(r2_train, 3))
    print("R2 TEST  :", round(r2_test, 3))
    print("RMSE TRAIN :", round(rmse_train, 3))
    print("RMSE TEST  :", round(rmse_test, 3))

    # Petite v√©rification d'overfitting
    if (r2_train - r2_test > 0.05) or (rmse_test > 2 * rmse_train):
        print("‚ö† Overfitting possible")
    else:
        print("‚úÖ Pas d'overfitting important")
```

```python

from sklearn.svm import SVR
from sklearn.metrics import r2_score

from hyperopt import hp, fmin, tpe, Trials, STATUS_OK

# --- Space Hyperopt ---
space = {
    "C": hp.loguniform("C", np.log(0.1), np.log(200)),
    "epsilon": hp.uniform("epsilon", 0.01, 0.2),
    "gamma": hp.choice("gamma", ["scale", 0.1, 0.01])
}

gamma_choices = ["scale", 0.1, 0.01]

def objective(params):
    scores = []

    # convert gamma index -> valeur (important!)
    gamma_val = params["gamma"]

    for train_idx, val_idx in tscv.split(X_train):
        X_tr, X_val = X_train.iloc[train_idx], X_train.iloc[val_idx]
        y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]

        model = Pipeline([
            ("scaler", StandardScaler()),
            ("model", SVR(
                kernel="rbf",
                C=params["C"],
                epsilon=params["epsilon"],
                gamma=gamma_val
            ))
        ])

        model.fit(X_tr, y_tr)
        preds = model.predict(X_val)
        scores.append(r2_score(y_val, preds))

    return {"loss": -np.mean(scores), "status": STATUS_OK}

trials = Trials()

best = fmin(
    fn=objective,
    space=space,
    algo=tpe.suggest,
    max_evals=20,
    trials=trials,
    rstate=np.random.default_rng(42)
)


best_params_svr = {
    "C": best["C"],
    "epsilon": best["epsilon"],
    "gamma": gamma_choices[best["gamma"]] if isinstance(best["gamma"], (int, np.integer)) else best["gamma"]
}



best_svr = Pipeline([
    ("scaler", StandardScaler()),
    ("model", SVR(
        kernel="rbf",
        C=best_params_svr["C"],
        epsilon=best_params_svr["epsilon"],
        gamma=best_params_svr["gamma"]
    ))
])

best_svr.fit(X_train, y_train)
print("Best SVR params:", best_params_svr)

```

```python
class HyperoptResult:
    def __init__(self, best_estimator, best_params):
        self.best_estimator_ = best_estimator
        self.best_params_ = best_params

svr_result = HyperoptResult(best_svr, best_params_svr)
```
