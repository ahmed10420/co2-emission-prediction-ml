```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler

# ===============================
# 1 D√©finir le target & dataset
# ===============================

target_col = "Taux de Co2"

df = df_clean.copy()

# enlever lignes
df = df.dropna()

X = df.drop(columns=[target_col])
y = df[target_col]

# ===============================
# 2  Split temporel (PAS shuffle)
# ===============================

n = len(df)
train_size = int(n * 0.8)

# s√©curit√© :
if train_size >= n:
    train_size = n - 1

X_train = X.iloc[:train_size]
X_test  = X.iloc[train_size:]

y_train = y.iloc[:train_size]
y_test  = y.iloc[train_size:]

print("Train samples =", len(X_train))
print("Test samples  =", len(X_test))

# ===============================
# 3Ô∏è‚É£  Normalisation
# ===============================

scaler_X = MinMaxScaler()
X_train = scaler_X.fit_transform(X_train)
X_test  = scaler_X.transform(X_test)

# ===============================
# 4Ô∏è‚É£  Mod√®le XGBoost
# ===============================

model_xgb = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=0
)

model_xgb.fit(X_train, y_train)

# ===============================
# 5Ô∏è‚É£  √âvaluation
# ===============================

y_pred = model_xgb.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("\n=========== RESULTS ===========")
print("RMSE =", rmse)
print("R¬≤   =", r2)
```

```python
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import r2_score, mean_squared_error
```

```python
X = X_selected
y = df_model["Taux de Co2"] # y is already defined, but re-assigning for clarity
```

```python
split_index = int(len(X) * 0.8)

X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]
```

```python
tscv = TimeSeriesSplit(n_splits=5)
```

```python
pipeline_ridge = Pipeline([
    ("scaler", StandardScaler()),
    ("model", Ridge())
])

param_grid_ridge = {
    "model__alpha": [0.1, 1, 10, 100]
}

grid_ridge = GridSearchCV(
    pipeline_ridge,
    param_grid_ridge,
    cv=tscv,
    scoring="r2"
)

grid_ridge.fit(X_train, y_train)
```

```python
param_grid_rf = {
    "n_estimators": [100, 200],
    "max_depth": [10, 20, None]
}

grid_rf = GridSearchCV(
    RandomForestRegressor(random_state=42),
    param_grid_rf,
    cv=tscv,
    scoring="r2",
    n_jobs=-1
)

grid_rf.fit(X_train, y_train)
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor(
    objective="reg:squarederror",
    random_state=42
)
param_grid_xgb = {
    "n_estimators": [100, 200, 500],
    "max_depth": [3, 4, 6],
    "learning_rate": [0.01, 0.05, 0.1],
    "subsample": [0.8, 1.0],
    "colsample_bytree": [0.8, 1.0]
}

# üîπ GridSearchCV avec TimeSeriesSplit (tscv d√©j√† d√©fini)
grid_xgb = GridSearchCV(
    xgb,
    param_grid_xgb,
    cv=tscv,
    scoring="r2",
    n_jobs=-1
)

grid_xgb.fit(X_train, y_train)
```

```python
import numpy as np

models = {
    "Ridge": grid_ridge,
    "Random Forest": grid_rf,
    "XGBoost": grid_xgb
}

for name, model in models.items():
    # Pr√©dictions sur TRAIN
    y_train_pred = model.predict(X_train)
    r2_train = r2_score(y_train, y_train_pred)
    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))

    # Pr√©dictions sur TEST
    y_test_pred = model.predict(X_test)
    r2_test = r2_score(y_test, y_test_pred)
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))

    print(f"\n{name}")
    print("Best params :", model.best_params_)
    print("R2 TRAIN :", round(r2_train, 3))
    print("R2 TEST  :", round(r2_test, 3))
    print("RMSE TRAIN :", round(rmse_train, 3))
    print("RMSE TEST  :", round(rmse_test, 3))

    # Petite v√©rification d'overfitting
    if (r2_train - r2_test > 0.05) or (rmse_test > 2 * rmse_train):
        print("‚ö† Overfitting possible")
    else:
        print("‚úÖ Pas d'overfitting important")
```

```python

from sklearn.svm import SVR
from sklearn.metrics import r2_score

from hyperopt import hp, fmin, tpe, Trials, STATUS_OK

# --- Space Hyperopt ---
space = {
    "C": hp.loguniform("C", np.log(0.1), np.log(200)),
    "epsilon": hp.uniform("epsilon", 0.01, 0.2),
    "gamma": hp.choice("gamma", ["scale", 0.1, 0.01])
}

gamma_choices = ["scale", 0.1, 0.01]

def objective(params):
    scores = []

    # convert gamma index -> valeur (important!)
    gamma_val = params["gamma"]

    for train_idx, val_idx in tscv.split(X_train):
        X_tr, X_val = X_train.iloc[train_idx], X_train.iloc[val_idx]
        y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]

        model = Pipeline([
            ("scaler", StandardScaler()),
            ("model", SVR(
                kernel="rbf",
                C=params["C"],
                epsilon=params["epsilon"],
                gamma=gamma_val
            ))
        ])

        model.fit(X_tr, y_tr)
        preds = model.predict(X_val)
        scores.append(r2_score(y_val, preds))

    return {"loss": -np.mean(scores), "status": STATUS_OK}

trials = Trials()

best = fmin(
    fn=objective,
    space=space,
    algo=tpe.suggest,
    max_evals=20,
    trials=trials,
    rstate=np.random.default_rng(42)
)


best_params_svr = {
    "C": best["C"],
    "epsilon": best["epsilon"],
    "gamma": gamma_choices[best["gamma"]] if isinstance(best["gamma"], (int, np.integer)) else best["gamma"]
}



best_svr = Pipeline([
    ("scaler", StandardScaler()),
    ("model", SVR(
        kernel="rbf",
        C=best_params_svr["C"],
        epsilon=best_params_svr["epsilon"],
        gamma=best_params_svr["gamma"]
    ))
])

best_svr.fit(X_train, y_train)
print("Best SVR params:", best_params_svr)

```

```python
class HyperoptResult:
    def __init__(self, best_estimator, best_params):
        self.best_estimator_ = best_estimator
        self.best_params_ = best_params

svr_result = HyperoptResult(best_svr, best_params_svr)
```
